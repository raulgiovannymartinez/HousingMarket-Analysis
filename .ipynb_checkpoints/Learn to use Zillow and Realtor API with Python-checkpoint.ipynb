```python
import requests

url = "https://realtor.p.rapidapi.com/properties/v2/list-sold"

querystring = {"sort":"sold_date","city":"New York City","offset":"0","state_code":"NY","limit":"200"}

headers = {
    'x-rapidapi-host': "realtor.p.rapidapi.com",
    'x-rapidapi-key': "58f6607abdmsh9b762be77994354p12aa2cjsne00cebd883ac"
    }

response = requests.request("GET", url, headers=headers, params=querystring)
```

```python
import json

# convert response to dictionary
data_dict = json.loads(response.text)
```

```python
data_dict.keys()
```

```python
import numpy as np
import pandas as pd

# features on first and second layer
features_first = ['property_id', 'year_built', 'list_date', 'last_update', 'prop_status', 'beds', 'baths_full', 
                  'garage', 'baths_half', 'sold_history', 'price']
features_second = {'address': ['line', 'city', 'state', 'postal_code', 'lat', 'lon', 'county', 'neighborhood_name'],
                    'lot_size': ['size', 'units'],
                    'building_size': ['size', 'units']}

# parse features from each property
data = []
for p in data_dict['properties']:
    
    # define current lists, col names only saved on last iteration
    curr_list = []
    col_names = []
    
    # first dict layer
    for f in features_first:
        col_names.append(f)
        
        try:
            curr_list.append(p[f])
        except:
            curr_list.append(np.nan)
    
    # second dict layer
    for s,val in features_second.items():
        for v in val:
            col_names.append(v)
            
            try:
                curr_list.append(p[s][v])
            except:
                curr_list.append(np.nan)
        
    # save as list of tuples
    data.append(tuple(curr_list))
```

```python
# create dataframe
df = pd.DataFrame(data, columns = col_names)
df
```
